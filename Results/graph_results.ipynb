```python
import re
def time_string_to_seconds(time_string: str):
    # Extract minutes and seconds using regex
    match = re.match(r'(\d+)m(\d+\.\d+)s', time_string)
    if match:
        minutes, seconds = map(float, match.groups())
        total_seconds = minutes * 60 + seconds
        return total_seconds
    
for i, df in enumerate(all_data):
    print(i)
    for key in ['Real Time', 'User Time', 'Sys Time']:
        df[key] = df[key].apply(time_string_to_seconds)
```
